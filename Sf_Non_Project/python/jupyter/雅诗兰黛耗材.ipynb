```python
import pandas as pd
```

```python
file_path = ['3月进销存.xlsx','4月进销存.xlsx', '5月进销存.xlsx', '6月进销存.xlsx']
```

```python
out_data = pd.DataFrame()
for file in file_path:
    data = pd.read_excel(file, sheet_name = '到货明细')
    data = data.drop(['品牌','系统编码','渠道','物料类型','物料明细','商品编码', '唛头名称','耗材名称','总入库'], axis =1 )
    data = data.set_index('料号')
    data = data.stack()
    data = data.reset_index()
    out_data = pd.concat([out_data, data], axis = 0)
out_data.columns = ['料号','入库日期','入库数量']
out_data.head()
```

```python
out_data.to_excel('总入库.xlsx',index = False)
```

```python
in_data = pd.DataFrame()
date = ['3月','4月','5月','6月']
for i in range(0,len(file_path)):
    data = pd.read_excel(file_path[i], sheet_name = '现有库存')
    data = data[['料号','期初库存','采购入库','其它入库','领用出库','其它出库','现有库存']]
    data['月份'] = date[i]
    in_data = pd.concat([in_data, data], axis = 0)
in_data.head()
```

```python
in_data.to_excel('总盘点.xlsx',index = False)
```
